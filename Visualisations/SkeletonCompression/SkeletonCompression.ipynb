```python
from pygel3d import graph, jupyter_display as jd
import matplotlib.pyplot as plt
import numpy as np
import os

SKEL_PATH = r"F:\Git\Bachelor2.0\Data\Skeletons"
```

```python
skel_size = []
```

```python
objcs = os.listdir(SKEL_PATH)
for k in range(2, 41, 2):
    name = f"skel_{k}.g"
    if name in objcs:
        path = os.path.join(SKEL_PATH, name)
        s = graph.load(path)
        skel_size.append([k, len(s.nodes())])
```

```python
data = np.array(skel_size)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.plot(data[:, 0], data[:, 1])
ax.set_xticks([x for x in range(2, 41, 2)])
ax.set_xlabel("Number of neighbors in original graph")
ax.set_ylabel("Number of nodes in skeleton")

for dat in data:
    ax.text(dat[0], dat[1], dat[1])

plt.savefig("Skeletonreductionexample")
```
