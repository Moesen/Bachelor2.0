```python
n_samples = 100
random_state = np.random.RandomState(1)
X, y = datasets.make_circles(n_samples=n_samples, random_state=random_state, noise=.01)
outer, inner = 0, 1
labels = np.full(n_samples, -1.)
labels[0] = outer
labels[-1] = inner
```

```python
label_prop = LabelPropagation()
label_prop.fit(X, labels)
output_labels = label_prop.transduction_
```

```python
plt.figure(figsize=(5, 5))
plt.subplot(1, 1, 1)

plt.scatter(X[y== outer, 0], X[y == outer, 1], color="darksalmon", marker=".", label="'forgotten' outer circle'")
plt.scatter(X[y== inner, 0], X[y == inner, 1], color="lightgreen", marker=".", label="'forgotten' inner circle")
plt.scatter(X[labels == outer, 0], X[labels == outer, 1], color="red", marker="s", lw=0, label="outer labeled", s=30)
plt.scatter(X[labels == inner, 0], X[labels==inner, 1], color="lime", marker="s", lw=0, label="inner labeled", s=30)


plt.legend(scatterpoints=1, shadow=False, loc="upper right")

# plt.subplot(1,2,2)
# output_label_array = np.asarray(output_labels)
# outer_numbers = np.where(output_label_array == outer)[0]
# inner_numbers = np.where(output_label_array == inner)[0]
# unknown_numbers = np.where(output_label_array == -1)[0]
# plt.scatter(X[outer_numbers, 0], X[outer_numbers, 1], color="navy", marker ="s", lw=0, s=10, label="outer learned")
# plt.scatter(X[inner_numbers, 0], X[inner_numbers, 1], color="c", marker ="s", lw=0, s=10, label="inner learned")
# plt.scatter(X[unknown_numbers, 0], X[unknown_numbers, 1], color="darkorange", marker ="s", lw=0, s=10, label="unlabeled")
# plt.legend(scatterpoints=1, shadow=False, loc="upper right")
```
