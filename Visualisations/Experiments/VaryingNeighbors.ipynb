```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
PATH = r"F:\Git\Bachelor2.0\Data\Experiments\NumberOfNeighbors"
```

```python
LP = np.genfromtxt(os.path.join(PATH, "test_lp_4.csv"), delimiter=",")
SRG = np.genfromtxt(os.path.join(PATH, "test_srg_3.csv"), delimiter=",")
print(LP)
print(SRG)
```

```python
fig, (ax, ax1) = plt.subplots(2, 1, figsize=(10, 10))
ax.plot(LP[:, 0], LP[:, 1])
ax.plot(LP[:, 0], LP[:, 2])
ax.legend(["Reconstruction", "Test"])
ax.set_title("LP Varying K results")
ax.set_xlabel("K")
ax.set_ylabel("Accuracy%")
ax.set_xticks([x for x in range(4, 41, 4)])
ax.set_ylim(np.min(SRG[:,[1,2]]), np.max(LP[:, [1,2]])+2)

ax1.plot(SRG[:, 0], SRG[:, 1])
ax1.plot(SRG[:, 0], SRG[:, 2])
ax1.legend(["Reconstruction", "Test"])
ax1.set_title("SRG Varying K results")
ax1.set_xlabel("K")
ax1.set_ylabel("Accuracy%")
ax1.set_xticks([x for x in range(4, 41, 4)])
ax1.set_ylim(np.min(SRG[:,[1,2]]), np.max(LP[:, [1,2]])+2)

for row in SRG:
    ax1.text(row[0], row[1], f"{int(row[3])}")

fig.suptitle("S=60.000, P=90%", fontsize=14)

plt.tight_layout()
plt.savefig("VaryingKResults")
```
