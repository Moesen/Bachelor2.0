```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean
```

```python
a = np.array([255, 255, 255, 0, 0, 0])
b = np.array([0, 0, 0, 255, 255, 255])
dist = euclidean(a, b)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle(f'Euclidan Distance = {dist:.4}', fontsize=15)
ax1.imshow(b.reshape(2, 3).T, cmap="gray")
ax1.set_title("Shape to the left")
ax2.imshow(a.reshape(2, 3).T, cmap="gray")
ax2.set_title("Shape to the right")
plt.tight_layout()
plt.savefig("shape_position.png")
```
